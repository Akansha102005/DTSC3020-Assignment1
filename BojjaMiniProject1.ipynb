## Part 2 — The Receipt & Loyalty Stars  *(25 points)*


```python
# Akansha Bojja — Part 2

# --- Order values ---
latte_price  = 4.50
latte_qty    = 3
muffin_price = 3.25
muffin_qty   = 2

# --- Calculations ---
subtotal = latte_price * latte_qty + muffin_price * muffin_qty
tax      = subtotal * 8.25 / 100
total    = subtotal + tax

# --- Print receipt with f-strings ---
print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax (8.25%): ${tax:.2f}")
print(f"Total: ${total:.2f}")

# --- Loyalty stars ---
total_whole  = int(total)          # truncate to whole dollars
stars_earned = total_whole // 5    # one star per complete $5
toward_next  = total_whole % 5     # leftover dollars toward next star

print(f"Stars earned: {stars_earned}")
print(f"Dollars toward next star: {toward_next}")

# --- Budget remaining ---
budget       = 25.00
money_left   = budget - total
print(f"Money left from budget: ${money_left:.2f}")

# // gives how many full groups of 5 fit into the total (the stars),
# and % gives what remains after those groups are removed (e.g. 20 = 4*5 + 0, so 4 stars and $0 leftover).
```

## Part 3 — Customer Data Cleanup & Order Code  *(20 points)*


```python
# Akansha Bojja — Part 3

raw_customer = "   maRIa gonZALEZ   "
slogan = "Eagle Beans: best coffee in town"

# --- 1. Lengths before and after cleanup ---
print(f"Length of raw_customer: {len(raw_customer)}")

clean_customer = raw_customer.strip().lower()   # remove surrounding spaces, lowercase everything
print(f"Length of clean_customer: {len(clean_customer)}")

# The lengths differ because raw_customer has leading and trailing spaces that strip() removes.

# --- 2. First and last characters ---
print(f"First character: {clean_customer[0]}")
print(f"Last character:  {clean_customer[-1]}")

# --- 3. Build the order code ---
order_code = clean_customer[:3].upper() + "-" + str(year_opened)
print(f"Order code: {order_code}")

# --- 4. Promo slogan ---
promo_slogan = slogan.replace("coffee", "espresso")
print(promo_slogan)

# --- 5. Announcement ---
print(f"Order ready for {clean_customer}! Your code is {order_code}.")
```

## Part 5 — The Weekly Prize Drawing  *(20 points)*


```python
# Akansha Bojja — Part 5

import random

customers = ["Maria", "Devon", "Aisha", "Carlos", "Lena", "Tyler"]
songs = ["Jazz Vibes", "Lo-Fi Beats", "Morning Acoustic", "Indie Mix", "Soft Piano"]

# --- 1. Set seed for reproducibility ---
random.seed(42)

# --- 2. Lucky number and discount roll ---
lucky_number   = random.randint(1, 50)
discount_roll  = random.random()
print(f"Lucky number: {lucky_number}")
print(f"Discount roll: {discount_roll:.4f}")

# --- 3. Prize drawing ---
free_drink_winner = random.choice(customers)
coupon_winners    = random.sample(customers, 2)
print(f"Free-drink winner: {free_drink_winner}")
print(f"Coupon winners: {coupon_winners[0]} and {coupon_winners[1]}")

# --- 4. Shuffle and print playlist ---
random.shuffle(songs)
print(f"New playlist order: {songs}")

# Running the cell twice gives the SAME winners every time because random.seed(42) resets
# the random number generator to a fixed starting point before each run.
# Without the seed line, the winners would be different each time the cell is run.
```
